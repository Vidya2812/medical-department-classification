```python
import pandas as pd
import re

file_path = "D:/patient_notes - patient_notes.csv"
data = pd.read_csv(file_path)

print("Original Data:")
print(data.head())
```

```python
columns_to_extract = ['pn_num', 'pn_history']
extracted_data = data[columns_to_extract]
```

```python
extracted_data.head()
```

```python
def clean_text(text):
    # Remove special characters, punctuation marks, HTML tags, etc.
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    
    # Convert text to lowercase
    cleaned_text = cleaned_text.lower()
    return cleaned_text
```

```python
# Apply text cleaning to 'Symptoms_Description' column
extracted_data['Cleaned_pn_history'] = extracted_data['pn_history'].apply(clean_text)
```

```python
extracted_data.head()
```

```python
missing_values = extracted_data.isnull().sum()
```

```python
# Handle missing or null values (if any)
# For example, if you want to remove rows with missing values:
extracted_data.dropna(inplace=True)
```

```python
# Display the preprocessed data
print("\nPreprocessed Data:")
extracted_data.head()
```

```python
import spacy
```

```python
# Load the English language model in spaCy
nlp = spacy.load("en_core_web_sm")
```

```python
descriptions = extracted_data['Cleaned_pn_history'].tolist()
```

```python
# Tokenize each description
tokenized_descriptions = []
for desc in descriptions:
    doc = nlp(desc)
    tokens = [token.text for token in doc]
    tokenized_descriptions.append(tokens)
```
